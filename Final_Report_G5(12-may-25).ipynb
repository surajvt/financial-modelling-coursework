## 1.2 Optimization Theory and Mathematical Formulas :
### 1.2.1 Portfolio Optimization

We used the ideas behind modern portfolio theory (MPT) to build and test different portfolios. Python’s optimization tools helped us figure out which asset combinations gave us the highest Sharpe ratio or the lowest risk, depending on the goal.
**Expected Return:**
$$
E(R_p) = \sum_{i=1}^{n} w_i \times E(R_i)
$$

**Portfolio Variance:**

$$
sigma_p^2 = \mathbf{w}^T \Sigma \mathbf{w}
$$

**Sharpe Ratio:**
$$
{Sharpe Ratio} = \frac{E(R_p) - R_f}{\sigma_p}
$$

- Optimization focused on maximizing Sharpe ratio while satisfying investment constraints.

## 1.3 Forecasted Value at Risk (VaR)
**Value-at-Risk (VaR):**
- To Estimate worst expected loss at the given confidence level.
**Approaches:**
- Historical Simulation: Uses actual past returns.
- Parametric (Variance-Covariance): Assumes returns are normally distributed.
- Monte Carlo Simulation: Random future scenarios drawn from statistical distribution:

$$
R_{simulated} = \mu + \sigma Z, \quad Z \sim N(0,1)
$$

## 1.4 GARCH Modeling and Stochastic Volatility

To track how market volatility changed over time, we used the `arch` library to fit GARCH(1,1) models. This helped us see which assets were more stable and which showed more risk over certain periods.
**GARCH(1,1) Model:**
$$
\sigma_t^2 = \omega + \alpha \epsilon_{t-1}^2 + \beta \sigma_{t-1}^2
$$

Where:
- \( \omega \): Long-term volatility component
- \( \alpha \): Short-term shocks
- \( \beta \): Persistence of volatility

**Why GARCH?**
- Captures volatility clustering—essential for realistic VaR calculations.
- Outperforms constant volatility models.

# **4.Normality-Tests**

````python
## Normality Tests Function Definition

'''
To supplement visual inspections (Q-Q plots), we define a function to statistically test the normality of asset return distributions.
This includes skewness, kurtosis, and the D’Agostino-Pearson omnibus test.
                                                                          '''


# Define a reusable function to test for normality in asset return distribution.
def normality_tests(arr):
    ''' Tests for normality distribution of given data set.
    
    Parameters
    ==========
    array: ndarray
        object to generate statistics on
    '''
    # Measure skewness and provide associated test p -value 
    print("Skew of data set  %14.3f" % scs.skew(arr))
    print("Skew test p-value %14.3f" % scs.skewtest(arr)[1])

    # Measure kurtosis and provide associated test p-value 
    print("Kurt of data set  %14.3f" % scs.kurtosis(arr))
    print("Kurt test p-value %14.3f" % scs.kurtosistest(arr)[1])
    
    # Omnibus test combining skew and kurtosis for normality  
    print("Norm test p-value %14.3f" % scs.normaltest(arr)[1])
````

````python
## Applying Normality Tests to All Assets

'''
Using a loop, we apply the previously defined `normality_tests` function to each asset's log return series.
This step ensures a consistent statistical evaluation of normality across all instruments in the portfolio.
                                                                                                        '''
# Loop through each asset column in the dataset 
for sym in data:
    print("\nResults for data %s" % sym)
    print(32 * "-")
    
 # Clean and extract log return series for the current asset
    log_data = np.array(log_returns[sym].dropna())

 # Apply the defined normality test function 
    normality_tests(log_data)
````

### Portfolio Risk in Numbers
Financial risk is conventionally quantified by  variance (SD)of a portfolio return (Hull, 2018).Portfolio variance is a sum of variance of its components and their covariances.

$$

$$
Or, in matrix notation:
$$
σ_p^2={w}^{T}∑{w}
$$
Where:
-$\\σ_p^2$: Varianz des Portfolios

- $\\Sigma$: The returnvariance/covariance matrix

The **standard deviation of a portfolio** (it’s a more “interpretable” measure of risk, since is in the same units as return) will be the square root of the variance:
$$
σ_p=\\sqrt{ \\mathbf{w^T} \\Sigma \\mathbf{w} }
$$

````python
# Defining Spline Functions
def f(x):
    ''' 
    Efficient-frontier-function (Spline Approx.). 
    This function takes an input x (volatility) and returns the 
    corresponding expected return using the B-spline representation defined earlier.
    '''    
    return sci.splev(x, tck, der=0) # Evaluate the spline at point x for the expected return
def df(x):
    ''' 
    The First derivative of an Efficient-Frontier-Function. 
    This function calculates rate of change of expected return with respect 
    to volatility (i.e., the slope of the efficient frontier at point x).
    '''
    return sci.splev(x, tck, der=1)#Evaluate the first derivative of the spline at x
````

````python
# Calculating the Capital Market Line and tangency portfolio.
# Defining the system of equations to find the tangency portfolio
def equations(p, rf=0.01):
    ''' 
    Equations to calculate the tangency portfolio on the Capital Market Line (CML).
    
    Parameters:
    - p: A tuple containing the following parameters:
        p[0] -> The expected return of the tangency portfolios
        p[1] -> Weight of the risky assets in tangency portfolios
        p[2] -> Volatility of tangency portfolios
    - rf: Risk-free rate (default is set to 0.01 or 1%)

    Returns:
    - A tuple of three equations that need to be satisfied.
    '''
    # Equation 1: Return of tangency portfolios should equal the risk-free rate
    eq1 = rf - p[0]
    # Equation 2: Expected returns of the tangency portfolio should equal the risk-free rate plus
    # the product of  weights of  risky asset and its volatility
    eq2 = rf + p[1] * p[2] - f(p[2]) # This ensures the expected return aligns with the efficient frontier
    eq3 = p[1] - df(p[2])
    return eq1, eq2, eq3 # Return the equations as a tuple for solving
````

````python
# calculating the optimal tangency portfolio using fsolve
opt = sco.fsolve(equations, [0.01, 0.5, 0.15]) #solve the equations with an initial guess
````

````python
# Portfolio Statistics
def portfolio_performance(weights, mean_returns, cov_matrix):
    """
    Calculate the expected portfolio returns and volatility.

    Args:
        weights (np.ndarray): Array of weights for the portfolio.
        mean_returns (np.ndarray): Array of mean returns for each asset.
        cov_matrix (np.ndarray): Covariance matrix of asset returns.

    Returns:
        tuple: A tuple containing the expected annual returns and the portfolio volatility (standard deviation).
    """
    # Calculate expected portfolio returns
    returns = np.dot(weights, mean_returns) * 252 # Annualize returns (assuming daily returns)

    # Calculate portfolio standard deviation (volatility)
    std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix * 252, weights)))  # Annualize volatility
    return returns, std # Return the expected returns and standard deviation
````

# **7. Forecasted value at Risk(VaR)**

## 7.1 Introduction

Value at Risk is a largely adopted technique in the financial risk management that quantifies the significant loss a portfolio could incur over a defined period,typical market conditions & with specified confidence-level.

In other words,it answers the question: *“What’s the worst loss I might face in a day (or week, or month), with, say, 95% confidence?”* This makes it a practical and powerful way to measure and communicate risk in simple monetary terms. It is used by investors, fund managers, banks, and regulators to guide financial decisions and manage exposure to market volatility.


---


### 7.2 Types of VaR

VaR can be calculated using different approaches, each suited to various situations based on data availability, portfolio complexity, and the assumptions one is willing to accept. The three primary techniques are as follows:


#### 1. Parametric (Variance-Covariance) VaR

The parametric approach is based on the concept that the returns of the portfolio are normally distributed and that classic bell-shaped curve. This makes it possible to estimate potential losses using a straightforward mathematical formula:

$$
\text{VaR} = z \cdot \sigma \cdot \sqrt{t}
$$

In this equation, the value of \( z \) corresponds to the confidence level selected—for example, 1.65 is typically used for 95% confidence. The symbol \( \sigma \) reflects the volatility or SD of the portfolio’s returns, while \( t \) refers to the length of the time horizon, such as 1 day or 10 days, depending on the analysis period.

This approach is widely favored due to its speed and simplicity, particularly when dealing with portfolios that exhibit normally distributed returns. However, it can underestimate risk in real-life situations where returns are not perfectly normal — such as during market crashes or periods of extreme volatility.


#### 2. Historical Simulation VaR

Historical simulation does not rely on any assumptions about the return distribution. Instead, it estimates future risk by analyzing actual historical return data.

To use this method, we gather daily return figures from the past, rank them from the lowest to the highest and identify the return value that aligns with the specified confidence level. As an example, to find out and calculate the 95% Value at Risk using 100 days of data, we would select the fifth worst-performing return.
That value is then treated as the estimated potential loss.

The main strength of this method is that it captures actual market movements, including rare but significant events. However, it operates under the assumption that future market behavior will mirror the past, which may not always be accurate.


#### 3. Monte Carlo Simulation VaR

The Monte Carlo method is more advanced and involves simulating thousands of possible future scenarios for portfolio returns. These simulations are generated using a statistical model, such as a normal distribution or a GARCH model that captures changing volatility.

After simulating many possible future outcomes, we calculate the loss for each and then find the loss that corresponds to the desired confidence level — typically the 5th percentile for 95% confidence.

This approach is expressed as:

$$
\text{VaR}_\alpha = \text{Quantile}_{1 - \alpha}(\text{Simulated Losses})
$$

Monte Carlo VaR is very flexible and works well for complex or non-linear portfolios, but it requires more time, data, and computing power. It’s especially useful when the portfolio includes options or instruments with uncertain outcomes.


````python
# Value at Risk
def calculate_var(weights, mean_returns, cov_matrix, confidence_level=0.95, horizon_days=30):
    portfolio_mean = np.dot(weights, mean_returns) * horizon_days
    portfolio_std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(horizon_days)
    z_score = 1.65  # 95% one-tailed
    var = -(portfolio_mean - z_score * portfolio_std)
    return var # Return the calculated Value at risk (VAR) for the portfolio
````

# **8. Stochastic Volatility and GARCH Model**

In financial markets, prices don’t move steadily — some days are calm, others are volatile. This shifting behavior is referred to as volatility, which often exhibits a pattern known as volatility clustering,where phases like high or low volatility tendencies continue for some time.

To model this behavior more realistically, we use models that allow volatility to change over time. Two popular ones are:

- Stochastic Volatility (SV) model
- "Garch" models (Generalized Autoregressive Conditional Heteroskedasticity)

These "models" are commonly applied in areas such as risk prediction, estimating Value at Risk (VaR), and pricing financial derivatives.


---

## 8.1 GARCH(1,1) Model

We are commonly using the GARCH model because it is practical and works well with financial time series data. While the name sounds complicated, the idea is straightforward: today’s volatility depends on yesterday’s volatility and on how surprising yesterday’s return was.

The GARCH(1,1) model is the most widely used version is the GARCH(1,1) model. The numbers (1,1) mean that the model looks at one lag of past squared returns (a shock) and one lag of past volatility to estimate current volatility.

For example:

- A return on any given day is made up of the average expected return & an unexpected component called shock.  
- The size of this shock depends on how volatile the market is.  
- Today’s volatility depends on how large yesterday’s shock was, and also on how volatile the market was yesterday.

This makes it well-suited to capture market behaviors like volatility clustering.


---

## 8.2 Key Components of GARCH(1,1)

#### 1. Return Equation

$$
r_t = \mu + \epsilon_t
$$

Where:  
\( r_t \): return at time \( t \)  
\( \mu \): average return  
\( \epsilon_t \): shock or innovation  

#### 2. Shock Equation

$$
\epsilon_t = \sigma_t z_t, \quad z_t \sim \mathcal{N}(0,1)
$$

Where:  
\( \sigma_t \): conditional standard deviation (volatility)  
\( z_t \): standard normal shock  

#### 3. Volatility Equation (Garch(1,1))

$$
\sigma_t^2 = \omega + \alpha \epsilon_{t-1}^2 + \beta \sigma_{t-1}^2
$$

Where:  
\( \omega > 0 \): base level of volatility  
\( \alpha \geq 0 \): reaction to past shocks (ARCH effect)  
\( \beta \geq 0 \): persistence of past volatility (GARCH effect)  

This means:  
- Large past shocks increase today’s volatility.  
- Persistent volatility leads to smoother volatility trends over time.

This methodology allows for dynamic conditional variance estimation, capturing how markets react differently to good and bad news (Abdalla, 2012). According to Almisshal and Emir (2021), this form reflects how new information and market memory influence current volatility — an essential feature of real market behavior.


---

## 8.3 Stochastic Volatility Model: Overview

In SV models, volatility is itself a random process, not directly determined by past returns.

1. Return model:

$$
r_t = \mu + \sigma_t z_t, \quad z_t \sim \mathcal{N}(0,1)
$$

Returns depend on time-varying volatility \( \sigma_t \), which changes independently.

2. Log-volatility evolution:

$$
\ln(\sigma_t^2) = \alpha + \phi \ln(\sigma_{t-1}^2) + \eta_t, \quad \eta_t \sim \mathcal{N}(0, \tau^2)
$$

This describes how the (log of) volatility evolves over time as an autoregressive process, with its own random shock \( \eta_t \).

---

## 8.4 Comparison: GARCH vs Stochastic Volatility

| Feature                   | GARCH                              | Stochastic Volatility (SV)         |
|--------------------------|------------------------------------|------------------------------------|
| Volatility source        | Deterministic function of past data | Latent stochastic process          |
| Estimation               | Maximum likelihood (efficient)     | Simulation-based (e.g., MCMC)      |
| Flexibility              | Less flexible, simpler to use       | More flexible, complex to estimate |
| Computational demand     | Low                                | High                               |
| Suitability for VaR      | Good for short-term VaR            | Suitable for complex structures and long horizons |

Summary:  
GARCH is simpler and widely used in practice.  
SV models can be more realistic but are harder to estimate.

---

## 8.5 Relation of GARCH and VaR

Accurate volatility modeling is critical in computing VaR, as VaR is directly proportional to volatility:

$$
\text{VaR}_{\alpha} = Z_{\alpha} \cdot \sigma_t \cdot \sqrt{t}
$$

Using models like GARCH or SV allows:

- Better capture of volatility clustering  
- More accurate forecasting of risk, especially during market stress  
- Dynamic, time-varying VaR estimates

In practice, GARCH models are more commonly used for VaR due to their ease of implementation and strong empirical performance.
